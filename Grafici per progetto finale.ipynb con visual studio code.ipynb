```python
import matplotlib.pyplot as plt

gruppi = ['PPE', 'SED', 'RE', 'VERDI', 'ID', 'ECR', 'GUE', 'NI', 'ALTRI']
variazioni = [3.85, -11.69, -31.48, -31.08, -20.55, 33.87, -4.88, -21.05, 0.0]

colori = ['green' if var > 0 else 'red' for var in variazioni]

plt.figure(figsize=(10, 6))
bars = plt.bar(gruppi, variazioni, color=colori)

plt.title('Variazione dei Seggi Parlamentari tra 2019 e 2024')
plt.xlabel('Gruppi Parlamentari')
plt.ylabel('Variazione Percentuale')
plt.xticks(rotation=45)

for bar, variazione in zip(bars, variazioni):
    height = bar.get_height()
    if variazione is not None:
        plt.text(bar.get_x() + bar.get_width() / 2, height, f'{variazione:.2f}%', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
